```python
class MaskDataset(Dataset):
    def __init__(self, df, img_dir, transform=None):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.transform = transform

        self.label_map = {
            "mask": 1,
            "no_mask": 0
        }

    def __len__(self):
        return len(self.df)

    def _normalize_label(self, raw_label):
        """
        Convert ANY label type to int {0,1}
        """
        # numpy / python numeric
        if isinstance(raw_label, (int, float, np.integer)):
            return int(raw_label)

        # string
        raw_label = str(raw_label).strip().lower()

        # string-number
        if raw_label.isdigit():
            return int(raw_label)

        # class name
        return self.label_map[raw_label]

    def __getitem__(self, idx):
        # image
        img_name = self.df.iloc[idx, 0]
        img_path = os.path.join(self.img_dir, img_name)
        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        # label (if exists)
        if self.df.shape[1] > 1:
            raw_label = self.df.iloc[idx, 1]
            label = self._normalize_label(raw_label)
            return image, label

        return image
```

```python
from sklearn.model_selection import train_test_split

train_df = pd.read_csv("train_labels.csv")

train_df, val_df = train_test_split(
    train_df,
    test_size=0.1,
    stratify=train_df["target"],
    random_state=42
)
```

```python
from torchvision import transforms
from PIL import ImageOps

class PadResize:
    def __init__(self, size, fill=0):
        self.size = size
        self.fill = fill

    def __call__(self, img):
        # pad image to square first
        w, h = img.size
        max_side = max(w, h)
        padded = ImageOps.pad(img, (max_side, max_side), method=Image.BICUBIC, color=self.fill)
        # resize to target
        return padded.resize((self.size, self.size))
```

```python
from torchvision import transforms

train_tfms = transforms.Compose([
     transforms.RandomHorizontalFlip(p=0.5),   # do_flip
    transforms.RandomVerticalFlip(p=0.5),     # flip_vert
    transforms.Resize((128,128)),
    transforms.RandomRotation(30),            # max_rotate
     transforms.ColorJitter(brightness=0.4*4, contrast=0.4*4, saturation=0.4*4, hue=0.1*4),  # max_lighting * mult
    transforms.RandomAffine(degrees=0, shear=20*4),  # max_warp * mult         # max_warp
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])

val_tfms = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
from torch.utils.data import DataLoader

train_ds = MaskDataset(train_df, "images/images", train_tfms)
val_ds = MaskDataset(val_df, "images/images", val_tfms)

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=32, shuffle=False)
```

```python
import timm
import torch

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

model = timm.create_model(
    "convnext_tiny",
    pretrained=True,
    num_classes=2
).to(DEVICE)
```

```python
from torchvision.transforms import v2
mixup = v2.MixUp(num_classes=2, alpha=0.4)
```

```python
import torch.nn as nn

criterion = nn.CrossEntropyLoss()

# 5 Optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-2, weight_decay=1e-2)
# 6 LR scheduler (optional but FastAI does this)
scheduler = torch.optim.lr_scheduler.OneCycleLR(
    optimizer, max_lr=3e-4,
    steps_per_epoch=len(train_loader),
    epochs=20
)
```

```python
# MixUp function (optional)
# ------------------------
def mixup_data(x, y, alpha=0.4):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1
    batch_size = x.size()[0]
    index = torch.randperm(batch_size).to(DEVICE)
    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# ------------------------
# Training loop
# ------------------------
EPOCHS = 20
USE_MIXUP = True

for epoch in range(EPOCHS):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in tqdm(train_loader):
        images = images.to(DEVICE)
        labels = labels.to(DEVICE)

        optimizer.zero_grad()

        if USE_MIXUP:
            images, y_a, y_b, lam = mixup_data(images, labels)
            outputs = model(images)
            loss = mixup_criterion(criterion, outputs, y_a, y_b, lam)
        else:
            outputs = model(images)
            loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()
        scheduler.step()

        running_loss += loss.item() * images.size(0)

        # Accuracy (hard labels, ignore MixUp effect)
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    train_loss = running_loss / total
    train_acc = correct / total

    # ------------------------
    # Validation
    # ------------------------
    model.eval()
    val_loss = 0.0
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(DEVICE)
            labels = labels.to(DEVICE)
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            val_correct += (predicted == labels).sum().item()
            val_total += labels.size(0)

    val_loss /= val_total
    val_acc = val_correct / val_total

    print(f"Epoch {epoch+1}/{EPOCHS} | "
          f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
          f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")
```

# Minimal Example - 10 Lines of Code

We've got a lot of explanation in here, but here is a minimal submission from scratch in only 10 lines of code. You'll notice we're using a fancier model, the Convnext tiny model from Timm. Is it better? You'll have to test and find out.

We're also adding .mixup() since that sounds fancy, and training for longer than the example. Feel free to play with this, and see how good you can get it!

You can read about mixup and friends [here](https://docs.fast.ai/callback.mixup.html#mixup)

Good luck :)


```python
from fastai.vision.all import * #import everthing from vision
import pandas as pd
import timm
```

```python
train_df = pd.read_csv("train_labels.csv")
submission = pd.read_csv("SampleSubmission.csv")
Nosemask = DataBlock(blocks=(ImageBlock, CategoryBlock), splitter=TrainTestSplitter(0.1, stratify=train_df["target"]),get_x = ColReader(0, pref = "images/images/" ),get_y=ColReader(1),item_tfms = Resize(460), batch_tfms = aug_transforms(do_flip=True,flip_vert=True,max_lighting=0.4,max_zoom=1.2,max_warp=0.2,max_rotate=30,xtra_tfms=None))
dls = Nosemask.dataloaders(train_df, bs=16, num_workers=0)
learn = vision_learner(dls,'convnext_tiny', metrics=[accuracy], path=".") #try convnext_base and convnext_large
learn.fine_tune(3,cbs=MixUp) #Apply Mixup #increase no of Epochs
tdl = dls.test_dl(submission) # quicly create test data loader
test_preds_tta, test_labels_tta = learn.get_preds(dl=tdl) #Get Preds
submission['target'] = [float(p[1]) for p in test_preds_tta]
submission.to_csv('convnext_tiny_model.csv', index=False)
```
